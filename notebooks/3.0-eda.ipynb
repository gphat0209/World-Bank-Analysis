## Making a question? (1.25 points)

After exploring the data, we have a better understanding of the data. Now let's see what questions can be answered with this data.

**A possible question is:** What is the population situation of the countries on the continent today?

**Answering this question will** help us to see the current population status and predictions about the future development of the labor force of the countries on the continent are under consideration.

**How we answer this question**: we will consider the Average Birth Rate and Death Rate for countries across all the year.

### Analyzing to get the answer

Step 01: Construct Average Birth Rate dataframe for each country.

```python
# Create a new dataframe
avg_birthrate_df = pd.DataFrame()

# Find average for each country
avg_birthrate_df["Avg.Birth Rate"] = cleaned_df.groupby("Country")["Birth Rate"].mean()

# Set the index as a column - countries
avg_birthrate_df["Country"] = avg_birthrate_df.index

# Drop the index
avg_birthrate_df.reset_index(drop=True, inplace=True)

# Sort the rows based of Average Birth Rate
avg_birthrate_df.sort_values("Avg.Birth Rate", inplace=True, ascending=False)

# Display
avg_birthrate_df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")
ax = sns.barplot(data=avg_birthrate_df.head(), x="Country", y= "Avg.Birth Rate")
```

Step 02: Construct Average Death Rate dataframe for each country.

```python
# Create a new dataframe
avg_deathrate_df = pd.DataFrame()

# Find average for each country
avg_deathrate_df ["Avg.Death Rate"] = cleaned_df.groupby("Country")["Death Rate"].mean()

# Set the index as a column - countries
avg_deathrate_df["Country"] = avg_deathrate_df .index

# Drop the index
avg_deathrate_df .reset_index(drop=True, inplace=True)

# Sort the rows based of Average Birth Rate
avg_deathrate_df .sort_values("Avg.Death Rate", inplace=True, ascending=False)

# Display
avg_deathrate_df .head()
```

```python
# Plotting for top countries
# YOUR CODE HERE
plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")
ax = sns.barplot(data=avg_deathrate_df.head(), x="Country", y= "Avg.Death Rate")
```

Guatemala, Haiti, Honduras Country, Bolivia, Belize are the countries with the highest average birth rate throughout the years. There is not much difference between the 5 countries.

Dominica, Haiti, Uruguay Country, St. Kitts and Nevis, Barbados are the countries with the highest average death rate throughout the years. There is not much difference between the 5 countries.

## Question 1: What is the unemployment situation of the countries in the continent?

### Analyzing to get the answer

```python
# YOUR CODE HERE
unemploy_df = pd.DataFrame()

unemploy_df["Average unemployment(%)"] = cleaned_df.groupby("Country")["Unemployment(%)"].mean()

unemploy_df["Country"] = unemploy_df.index

unemploy_df.reset_index(drop=True, inplace=True)

unemploy_df.sort_values("Average unemployment(%)", inplace=True, ascending=False)

unemploy_df.head()
```

**Observation:**

St. Lucia comes in second with about 18%.

Haiti Country, Guyana, Colombia are relatively similar in unployment rate (about 12.5%).


## Question 2: What is the male and female population in the countries on the continent?

### Analyzing to get the answer

```python
male_female_df = pd.DataFrame()
male_female_df["Average male population"] = cleaned_df.groupby("Country")["Male Population"].mean()
male_female_df["Average female population"] = cleaned_df.groupby("Country")["Female Population"].mean()
male_female_df["Average population"] = cleaned_df.groupby("Country")["Total Population"].mean()

male_female_df["Country"] = male_female_df.index

male_female_df.reset_index(drop=True, inplace=True)

male_female_df.sort_values("Average population", inplace=True, ascending=False)

male_female_df.head()

```

```python
X_axis = np.arange(5)
plt.bar(X_axis - 0.2, male_female_df.head(5)["Average female population"], 0.4, label = 'Female') 
plt.bar(X_axis + 0.2, male_female_df.head(5)["Average male population"], 0.4, label = 'Male') 


plt.xticks(X_axis, male_female_df.head(5)["Country"]) 
plt.xlabel("Country") 
plt.ylabel("Population count") 
plt.title("Male and female comparison") 
plt.legend() 
plt.show()
```

**Observation:**
In countries with highest population, female and male are relatively similar in population. But female is slightly higher in every country in the top 5.

## Question 3: What are the employment rate on agriculture and industry in countries with highest GDP?

```python
Q3_df = pd.DataFrame()
Q3_df["Average agriculture employment"] = cleaned_df.groupby("Country")["Employment in Agriculture(%)"].mean()
Q3_df["Average industry employment"] = cleaned_df.groupby("Country")["Employment in Industry(%)"].mean()
Q3_df["Average GDP in USD"] = cleaned_df.groupby("Country")["GDP in USD"].mean()

Q3_df["Country"] = Q3_df.index

Q3_df.reset_index(drop=True, inplace=True)

Q3_df.sort_values("Average GDP in USD", inplace=True, ascending=False)

Q3_df.head()
```

```python
X_axis = np.arange(5)
plt.bar(X_axis - 0.2, Q3_df.head(5)["Average agriculture employment"], 0.4, label = 'Agriculture') 
plt.bar(X_axis + 0.2, Q3_df.head(5)["Average industry employment"], 0.4, label = 'Industry') 


plt.xticks(X_axis, male_female_df.head(5)["Country"]) 
plt.xlabel("Country") 
plt.ylabel("Population count") 
plt.title("Industry and Agriculture") 
plt.legend() 
plt.show()
```

**Observation:**
In top 5 countries with highest GDP, the employment rates in industry are higher than those of agriculture.

In countries like United States, Mexico Country, the employment rates in industry are almost 5 times the rate in agriculture.

## Question 4: Does a country have a high GDP also have national income per capita?

```python
Q4_df = pd.DataFrame()
Q4_df["Average GDP"] = cleaned_df.groupby("Country")["GDP in USD"].mean()

Q4_df["Country"] = Q4_df.index

Q4_df.reset_index(drop=True, inplace=True)

Q4_df.sort_values("Average GDP", inplace=True, ascending=False)


plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")
ax = sns.barplot(data=Q4_df.head(), x="Country", y= "Average GDP")
```

```python
Q4_2_df = pd.DataFrame()
Q4_2_df["Average National Income per Capita"] = cleaned_df.groupby("Country")["National Income per Capita"].mean()

Q4_2_df["Country"] = Q4_2_df.index

Q4_2_df.reset_index(drop=True, inplace=True)

# Q4_2_df.sort_values("Average National Income per Capita", inplace=True, ascending=False)
us_value = Q4_2_df.loc[Q4_2_df['Country'] == 'United States']["Average National Income per Capita"]
brazil_value = Q4_2_df.loc[Q4_2_df['Country'] == 'Brazil']["Average National Income per Capita"]
mex_value = Q4_2_df.loc[Q4_2_df['Country'] == 'Mexico']["Average National Income per Capita"]
bri_value = Q4_2_df.loc[Q4_2_df['Country'] == 'British Virgin Islands']["Average National Income per Capita"]


country = ['United States', 'Brazil', 'Mexico', 'British Virgin Islands']
income_val = [us_value, brazil_value, mex_value, bri_value]

y_value = []
for x in income_val:
    x = str(x)
    x = x[5:9]
    y_value.append(float(x))

fig = plt.figure(figsize = (10, 5))

plt.bar(country, y_value, color ='maroon', 
        width = 0.4) 
plt.xlabel("Countries")
plt.ylabel("Average National Income per Capita")
plt.show()
```

**Observation:**
United States, Brazil, Mexico, British Virgin Islands are in top 5 countries with highest average GDP.

And similarly, they also have high average national income per capita (except for Mexico).
